```python
import pickle
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import os
import pandas as pd

import warnings
warnings.simplefilter("ignore")
```

```python
## Params
# psi: [0, 1, 2]
# Wc : [0, 0.5, 1]
# dens: [0.0125, 0.05, 0.2, 0.55]
# D : [0.5, 0.1, 1. ]
df = pd.read_json('df_dist_sims_09_23.JSON')
df.head()
```

```python
plt.rcParams['figure.figsize'] = 15, 8
fig,axs = plt.subplots(3,4)
i = 0
a = axs.ravel()
# diffusion rate
#for D, gg_D in df.groupby('D'):

# adhesion energy
for Wc, gg_Wc in df.groupby('Wc'):
    # density
    for sr, gg_sr in gg_Wc.groupby('density'):
        legend = []
        # CIL
        for psi, gg_psi in gg_sr.groupby('psi'):
            gg_plot = gg_psi[(gg_psi.time==990)]
            a[i].hist(gg_plot['size'], bins=10, log=True, alpha=0.5)
            legend.append(r'$\psi=%0.2g$, $\phi=%0.2g$, Wc=%0.2g'%(psi,sr, Wc))
        axs[i // 4][i % 4].legend(legend)
        # busca un mejor límite superior, parece que 10^2 es bueno
        #axs[i // 4][i % 4].set_ylim([0,150])
        i += 1
        
plt.suptitle('Cluster size, all Dr')
plt.tight_layout()
```

```python
## Params
# psi: [0, 0.5, 1, 1.5, 2]
# Wc : [0, 0.5, 1]
# dens: [0.0125, 0.05, 0.2, 0.55]
# D : [0.5, 0.1, 1. ]

df = pd.read_json('df_dist_sims_09_24.JSON')
df.head()
```

```python
# All diffusion rates
plt.rcParams['figure.figsize'] = 30, 16
fig,axs = plt.subplots(4,5)
i = 0
a = axs.ravel()
# density
for sr, gg_sr in df.groupby('density'):
    # CIL
    for psi, gg_psi in gg_sr.groupby('psi'):
        legend = []
        # adhesion energy
        for Wc, gg_Wc in gg_psi.groupby('Wc'):
            gg_plot = gg_Wc[(gg_Wc.time==990)]
            a[i].hist(gg_plot['size'], bins=10, log=True, alpha=0.5)
            legend.append(r'$\psi=%0.2g$, $\phi=%0.2g$, Wc=%0.2g'%(psi,sr, Wc))

            #legend.append(r'$\psi=%0.2g$, $\phi=%0.2g$, D=%0.2g, Wc=%0.2g'%(psi,sr, D, Wc))
        axs[i // 5][i % 5].legend(legend)
        # busca un mejor límite superior, parece que 10^2 es bueno
        axs[i // 5][i % 5].set_ylim([0,1e4])
        i += 1

plt.suptitle(f"Cluster size, all Dr")
plt.tight_layout()
plt.plot()
```

```python
# diffusion rate 0.1
D = 0.1
print(f"Difussion rate: {D}")
gg_D = df[df.D==D]
plt.rcParams['figure.figsize'] = 30, 16
fig,axs = plt.subplots(4,5)
i = 0
a = axs.ravel()
# density
for sr, gg_sr in gg_D.groupby('density'):
    # CIL
    for psi, gg_psi in gg_sr.groupby('psi'):
        legend = []
        # adhesion energy
        for Wc, gg_Wc in gg_psi.groupby('Wc'):
            gg_plot = gg_Wc[(gg_Wc.time==990)]
            a[i].hist(gg_plot['size'], bins=10, log=True, alpha=0.5)
            legend.append(r'$\psi=%0.2g$, $\phi=%0.2g$, Wc=%0.2g'%(psi,sr, Wc))

            #legend.append(r'$\psi=%0.2g$, $\phi=%0.2g$, D=%0.2g, Wc=%0.2g'%(psi,sr, D, Wc))
        axs[i // 5][i % 5].legend(legend)
        # busca un mejor límite superior, parece que 10^2 es bueno
        axs[i // 5][i % 5].set_ylim([0,1e4])
        i += 1

plt.suptitle(f"Cluster size, D = {D}")
plt.tight_layout()
plt.plot()
```
